```python
#正则表达式应用

# 正则表达式爬虫

import re
a = 'Hello 123 world'
result = re.findall('\d\d\d', a)
print(result)

import re
content = 'Hello 123 world 456 华晓智Python基础教学135'
result  = re.findall('\d\d\d',content)
print(result)


# 非贪婪匹配之 "(.*?)"
import re
res = '文本A百度新闻文本B'
source = re.findall('文本A(.*?)文本B', res)
print(source)

import re
res = '文本A百度新闻文本B,新闻标题文本A新浪财经文本B, 文本A搜狐新闻文本B新闻网址'
source = re.findall('文本A(.*?)文本B', res)
print(source)


# 非贪婪匹配之 ".*?"
import re
res = '<h3>文本C<变化的网址>文本D新闻标题</h3>'
p_title = '<h3>文本C.*?文本D(.*?)</h3>'
title  = re.findall(p_title, res)
print(title)

import re
res = '<h3 class="news-title_1YtI1">网址及一堆不关心的内容><!- - s - text- ->双十一点燃线下经济“小时达”服务或成<em>阿里巴巴</em>增长新引擎<!- -/s-text- -></a>'
p_title = '<h3 class="news-title_1YtI1">.*?>(.*?)</a>'
title = re.findall(p_title, res)
print(title)

# (.*?)本身无法匹配换行符\n，自动考虑换行的修饰符 re.S
import re
res = '''文本A
         百度新闻文本B'''
p_source = '文本A(.*?)文本B'
source   = re.findall(p_source, res, re.S)
print(source)

# 清洗信息
title1 = '<!- - s - text- ->双十一点燃线下经济“小时达”服务或成<em>阿里巴巴</em>增长新引擎<!- - s - text- ->'
title1 =title1.replace('<!- - s - text- ->','')
title1 =title1.replace('<em>','')
title1 =title1.replace('</em>','')
print(title1)

#sub函数
title2 = '<!- - s - text- ->双十一点燃线下经济“小时达”服务或成<em>阿里巴巴</em>增长新引擎<!- - s - text- ->'
title2 = re.sub('<.*?>','', title2)
print(title2)

#[]的用法，因为[]只能指代一个字符，所以[]中的'*' '+' '?'等均指代其本身符号自己
company = '*华能信托'
company1 = re.sub('[*]','', company) 
print(company1)
```

```python
# 证券日报数据爬取-正则表达式
import requests
import re

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}

# 1.获取网页源代码
url = 'http://search.zqrb.cn/search.php?src=all&q=贵州茅台&f=_all&s=newsdate_DESC'
res = requests.get(url, headers=headers).text
# print(res)

# 2.解析网页源代码提取信息
p_title = '<a href=".*?" target="_blank"><h4>(.*?)</h4></a>'
p_href = '<a href="(.*?)" target="_blank"><h4>.*?</h4></a>'
p_date = '<span><strong>时间:</strong>(.*?)</span>'
title = re.findall(p_title, res)
href = re.findall(p_href, res)
date = re.findall(p_date, res)

# 3.清洗 & 打印数据
for i in range(len(title)):
    title[i] = re.sub('<.*?>', '', title[i])  # 有的标题里有<em></em>这样的标签，这里清洗下数据
    date[i] = date[i].split(' ')[0]           # 日期只需要2020-11-18 06:15:27的年月日信息
    print(title[i] + '(' + '证券日报' + ' ' + date[i] + ')')  # 提取的标题自带序号
    print(href[i])

  
```

```python
#豆瓣电影top250 海报图片爬虫和保存到当地
import requests
import re

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}
# 1.获取网页源代码
url = 'https://movie.douban.com/top250'
res = requests.get(url, headers=headers).text
# print(res)

# 2.通过正则表达式获取名称title和图片网址img
p_title = '<img width="100" alt="(.*?)"'
title = re.findall(p_title, res)
p_img = '<img width="100" alt=".*?" src="(.*?)"'
img = re.findall(p_img, res)

# 3.打印图片名称，并通过requests库下载图片
for i in range(len(title)):
    print(str(i + 1) + '.' + title[i])
    print(img[i])
    res = requests.get(img[i])  # 下面开始下载图片
    file = open('D:/images/' + title[i] + '.png', 'wb')  # 需要提前创建好images文件夹！
    file.write(res.content)
    file.close()
```

```python
import requests
import re
from urllib.request import urlretrieve   # 也可以使用urlretrieve()批量下载

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}


def db(page):
    # 1.获取网页源代码
    num = (page - 1) * 25  # 页面参数规律是（页数-1）* 25
    url = 'https://movie.douban.com/top250?start=' + str(num)
    res = requests.get(url, headers=headers).text

    # 2.通过正则表达式获取名称title和图片网址img
    p_title = '<img width="100" alt="(.*?)"'
    title = re.findall(p_title, res)
    p_img = '<img width="100" alt=".*?" src="(.*?)"'
    img = re.findall(p_img, res)

    # 3.打印图片名称，并通过requests库下载图片，这里urlretrieve是requests库中的一个快捷函数，帮助我们快速从网页下载图片并保存到本地
    for i in range(len(title)):
        print(str(i + 1) + '.' + title[i])
        print(img[i])
        urlretrieve(img[i], 'D:/images/' + title[i] + '.png')  # 需提前创建好images文件夹


for i in range(2):  # 注意i是从0开始的序号，所以下面要写i+1页
    db(i+1)
    print('第' + str(i+1) + '页爬取成功！')
```
